### 6. Linear SVM (1 point)
Use cross-validation to train and optimize the hyperparameters for a linear support vector machine. Use the same technique as before.

Use the following grid:

Note that we're choosing relatively small values for `C`. This is allowed because our data is normalized.

Save the grid results in `grid_search`. Save the best classifier in `linear_svm_classifier`. There are many ways to create a linear SVM classifier. Look up the `sklearn` docs to choose the fastest one (in terms of performance).

```python
parameters_LSVM={
    "C":[0.1, 0.5, 0.8, 1, 1.5, 2, 6, 10, 15, 20],
      
}
```

```python
i = 1
for train_index,test_index in k_fold5.split(bank_attributes_train,bank_labels_train):
#     print("TRAIN:", type(train_index), "TEST:", test_index[0:2])
    x_train = bank_attributes_train.iloc[train_index]
    x_test = bank_attributes_train.iloc[test_index]
    y_train = bank_labels_train.iloc[train_index]
    y_test = bank_labels_train.iloc[test_index]
    grid_search5 = GridSearchCV(SVC(kernel='linear'),param_grid=parameters_LSVM,cv=5)    
    #Train the model
    grid_search5.fit(x_train, y_train) 
    print(i)
    print(grid_search5.best_params_)
    print(grid_search5.best_score_)
#     print(f"Accuracy for the fold no. {i} on the test set: {f1_score(y_test, grid_search.predict(x_test))}")
    i += 1
```
